```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import display
from PIL import Image
import numpy as np

imgpath = []
img = []
titles = []
images = []

imgpath.append("ref/tobey.jpg")
imgpath.append("ref/Kemas Muhammad Irsan Riza.jpg")
imgpath.append("ref/ridwan kamil.jpg")
imgpath.append("ref/Meza-Silvana.jpg")
imgpath.append("ref/revalina.jpg")
imgpath.append("ref/Putri Marino.jpg")

rows = 45
columns = 3
fig = plt.figure(figsize=(30, 400))

for x in range(3):
    
    img.append(cv2.cvtColor(cv2.imread(imgpath[x*2], cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
    img.append(cv2.cvtColor(cv2.imread(imgpath[x*2+1], cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
    titles.append(imgpath[x*2])
    titles.append(imgpath[x*2+1])
    images.append(img[x*2])
    images.append(img[x*2+1])
    
    plt.rcParams.update({'font.size': 30})

    fig.add_subplot(rows,columns,x*3+1)
    plt.imshow(images[x*2])
    plt.xlabel(titles[x*2])
    plt.xticks([])
    plt.yticks([])

    fig.add_subplot(rows,columns,x*3+2)
    plt.text(0.5, 0.5, "Hello World!", ha='center', va='center')
    plt.axis('off')

    fig.add_subplot(rows,columns,x*3+3)
    plt.imshow(images[x*2+1])
    plt.xlabel(titles[x*2+1])
    plt.xticks([])
    plt.yticks([])



plt.show()  
```
