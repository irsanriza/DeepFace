Menjalankan fungsi find dengan konfigurasi tertentu untuk setiap sumber data 

```python
from deepface import DeepFace
import os
import numpy as np

ref_path = "ref"
test_path = "test2"
ref = os.listdir(ref_path)
models = ["VGG-Face", "Facenet", "Facenet512", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib", "SFace"]
metrics = ["cosine", "euclidean", "euclidean_l2"]
backends = ['opencv', 'ssd', 'dlib', 'mtcnn', 'retinaface', 'mediapipe']
hasil = []
arrsumber = []
arrtujuan = []
arrhasil = []
arrdistance = []
i=0
arrx = np.empty((0,7))

for sumber in ref:
    match = DeepFace.find(img_path = ref_path + "/" + sumber, db_path = test_path, enforce_detection = False, model_name = models[2], distance_metric = metrics[2], detector_backend = backends[4])
    i+=1
    if match.size > 0:
        arrx = np.append(arrx, np.array([[i, Image.open(ref_path + "/" + sumber), sumber, Image.open(match["identity"][0]), match["identity"][0], "Benar", match["Facenet512_euclidean_l2"][0]]]), axis = 0)
        hasil.append([i, sumber, match["identity"][0], "Benar", match["Facenet512_euclidean_l2"][0]])
        arrsumber.append(sumber)
        arrtujuan.append(match["identity"][0])
        arrhasil.append("Benar")
        arrdistance.append(match["Facenet512_euclidean_l2"][0])
    elif match.size == 0:
        arrx = np.append(arrx, np.array([[i, Image.open(ref_path + "/" + sumber), sumber, "unknown", "unknown", "Salah", "0"]]), axis = 0)
        hasil.append([i, sumber, "unknown", "Salah", "0"])
        arrsumber.append(sumber)
        arrtujuan.append("Unknown")
        arrhasil.append("Salah")
        arrdistance.append("0")
```

Menampilkan hasil matching wajah untuk setiap foto di folder "ref"

```python
from tabulate import tabulate
header = ['No', 'Sumber', 'Tujuan', 'Hasil', 'Distance']
print(tabulate(hasil, headers=header))
```

```python
from tabulate import tabulate
header = ['No', 'gambar sumber', 'Nama Sumber', 'gambar tujuan', 'Gambar Tujuan', 'Hasil', 'Distance']
print(tabulate(arrx, headers=header))
#print(arrx)
```

```python
from IPython.display import display
from PIL import Image

for x in range(0, i):
    print(x+1, ". ", arrsumber[x])
    tmp_gambar = Image.open(ref_path + "/" + arrsumber[x])
    display(tmp_gambar.resize((int(tmp_gambar.width * 100 / tmp_gambar.height), 100), Image.NEAREST))
    print("gambar tujuan: ", arrtujuan[x], "\t ", arrhasil[x])
    if arrtujuan[x] != "Unknown":
        tmp_gambar = Image.open(arrtujuan[x])
        display(tmp_gambar.resize((int(tmp_gambar.width * 100 / tmp_gambar.height), 100), Image.NEAREST))
        print("distance: ", arrdistance[x])
    print("________________________________________")
```
